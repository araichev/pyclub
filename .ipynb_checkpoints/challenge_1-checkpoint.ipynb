Challenge 1
============
Take the Portland GTFS feed and create a CSV file that summarize trip data with the following columns:

- trip_id
- direction_id
- route_id
- start_time: first departure time of the trip
- end_time: last departure time of the trip
- duration: duration of the trip in hours
- start_stop_id: stop ID of the first stop of the trip
- end_stop_id: stop ID of the last stop of the trip
- num_stops: number of stops on trip
- distance: distance travelled by the trip in km

Note that the Portland feed distances are in feet.


```python
import sys
import os
import zipfile
import shutil
import datetime as dt
from collections import OrderedDict

import pandas as pd
import numpy as np

#home = '/home/araichev/pyclub/'
home = '/Users/araichev/pyclub/'
sys.path.append(home)
```

```python
# Define some utility functions
def time_it(f):
    """
    Wrapper for timing functions
    """
    def wrap(*args, **kwargs):
        t1 = dt.datetime.now()
        print('Timing', f.__name__)
        print(t1, 'Began process')
        result = f(*args, **kwargs)
        t2 = dt.datetime.now()
        minutes = (t2 - t1).seconds/60
        print(t2, 'Finished in %.2f min' % minutes)    
        return result
    return wrap

def seconds_to_timestr(seconds, inverse=False):
    """
    Return the given number of integer seconds as the time string '%H:%M:%S'.
    If ``inverse == True``, then do the inverse operation.
    In keeping with GTFS standards, the hours entry may be greater than 23.
    """
    if not inverse:
        try:
            seconds = int(seconds)
            hours, remainder = divmod(seconds, 3600)
            mins, secs = divmod(remainder, 60)
            result = '{:02d}:{:02d}:{:02d}'.format(hours, mins, secs)
        except:
            result = None
    else:
        try:
            hours, mins, seconds = seconds.split(':')
            result = int(hours)*3600 + int(mins)*60 + int(seconds)
        except:
            result = None
    return result

def to_km(x, units='km'):
    """
    Given a distance ``x`` in units ``units``,
    convert it to kilometers and return the result.
    """
    if units == 'km':
        return x
    if units == 'm':
        return x/1000
    if units == 'mi':
        return x*1.6093
    if units == 'ft':
        return x*0.00030480
```

```python
# Portland GTFS feed
path = 'data/challenge_1/portland_gtfs.zip'

# Unzip
if zipfile.is_zipfile(path):
    # Extract to temporary location
    zipped = True
    archive = zipfile.ZipFile(path)
    tmp_path = path.rstrip('.zip') + '/'
    archive.extractall(tmp_path)
    
# Read relevant data into Pandas data frames
trips = pd.read_csv(tmp_path + 'trips.txt', dtype={'route_id': str,
'trip_id': str, 'service_id': str, 'shape_id': str, 'stop_id': str})
stop_times = pd.read_csv(tmp_path + 'stop_times.txt', dtype={'stop_id': str,
'trip_id': str})

# Convert distances to km
stop_times['shape_dist_traveled'] = stop_times['shape_dist_traveled'].map(
  lambda x: to_km(x, 'ft'))

# Remove extracted directory
shutil.rmtree(tmp_path)
```

```python
# A quick look at the data
print(trips.head())
print(stop_times.head())
print('num stop times =', stop_times.shape[0])
```

```python
@time_it
def get_trips_stats(trips, stop_times):
    """
    Return a Pandas data frame with the following columns:

    - trip_id
    - direction_id
    - route_id
    - start_time: first departure time of the trip
    - end_time: last departure time of the trip
    - duration: duration of the trip in hours
    - start_stop_id: stop ID of the first stop of the trip 
    - end_stop_id: stop ID of the last stop of the trip
    - num_stops: number of stops on trip
    - distance: distance traveled by trip in km
    """
    num_trips = trips.shape[0]

    # Initialize data frame. Base it on trips.txt.
    stats = trips[['route_id', 'trip_id', 'direction_id']]

    # Compute start time, end time, duration
    f = pd.merge(trips, stop_times)

    # Convert departure times to seconds past midnight, 
    # to compute durations below
    f['departure_time'] = f['departure_time'].map(
      lambda x: seconds_to_timestr(x, inverse=True))
    g = f.groupby('trip_id')
    h = g['departure_time'].agg(OrderedDict([('start_time', np.min), 
      ('end_time', np.max)]))
    h['duration'] = (h['end_time'] - h['start_time'])/3600

    # Compute start stop and end stop
    def start_stop(group):
        i = group['departure_time'].argmin()
        return group['stop_id'].at[i]

    def end_stop(group):
        i = group['departure_time'].argmax()
        return group['stop_id'].at[i]

    h['start_stop'] = g.apply(start_stop)
    h['end_stop'] = g.apply(end_stop)
    h['num_stops'] = g.size()

    # Convert times back to time strings
    h[['start_time', 'end_time']] = h[['start_time', 'end_time']].\
      applymap(lambda x: seconds_to_timestr(int(x)))

    # Compute distances using shape_dist_traveled column
    h['distance'] = g.apply(lambda group: 
    group['shape_dist_traveled'].max())

    stats = pd.merge(stats, h.reset_index()).sort(['route_id', 
      'direction_id', 'start_time'])
    return stats
```

```python
# Run our main function and time it
#f = time_it(get_trips_stats)
trips_stats = get_trips_stats(trips, stop_times)
trips_stats
```

```python
# Write to CSV
trips_stats.to_csv('data/challenge_1/trips_stats.csv', index=False)

# Check the output (using some iPython magic)
%less data/challenge_1/trips_stats.csv 
```
